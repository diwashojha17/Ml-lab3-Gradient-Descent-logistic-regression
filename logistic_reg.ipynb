```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/content/drive/MyDrive/ML/dataset/diabetes.csv')
```

```python
df=df.head()
print(df)
```

```python
df=pd.DataFrame(df)
print(df)
```

```python
x = df[['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness',
        'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age']]
y = df['Outcome']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.4,random_state=42)
from sklearn.linear_model import LogisticRegression

model=LogisticRegression(max_iter=1000)
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

```python
#  visualize relationship between Glucose and Outcome
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8,5))
sns.regplot(x='Glucose', y='Outcome', data=df, logistic=True, ci=None, scatter_kws={'color':'blue'}, line_kws={'color':'red'})
plt.xlabel('Glucose Level')
plt.ylabel('Diabetes Outcome')
plt.title('Logistic Regression Curve for Diabetes Prediction')
plt.show()
```
